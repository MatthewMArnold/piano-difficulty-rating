```python


rep_paths = [
    "results/approach_windowed/results_rep_note_w9.json",
    "results/approach_windowed/results_rep_finger_nakamura_w9.json",
    "results/approach_windowed/results_rep_finger_w9.json",
    "results/approach_windowed/results_rep_prob_w9.json",
    "results/approach_windowed/results_rep_velocity_w9.json"
]




rep_p_paths = [
    "results/approach_windowed/step2pearson_results_rep_finger_nakamura_w9.json",
    "results/approach_windowed/step2pearson_results_rep_finger_w9.json",
    "results/approach_windowed/step2pearson_results_rep_note_w9.json",
    "results/approach_windowed/step2pearson_results_rep_prob_w9.json",
    "results/approach_windowed/step2pearson_results_rep_velocity_w9.json"
]

ws_p_paths = [
    "results/approach_windowed/step2pearson_results_rep_velocity_w5.json",
    "results/approach_windowed/step2pearson_results_rep_velocity_w9.json",
    "results/approach_windowed/step2pearson_results_rep_velocity_w13.json",
    "results/approach_windowed/step2pearson_results_rep_velocity_w19.json",
    "results/approach_windowed/step2pearson_results_rep_velocity_w27.json"
]

rep_c_paths = [
    "results/approach_windowed/step2rank_corr_results_rep_note_w9.json",
    "results/approach_windowed/step2rank_corr_results_rep_finger_nakamura_w9.json",
    "results/approach_windowed/step2rank_corr_results_rep_finger_w9.json",
    "results/approach_windowed/step2rank_corr_results_rep_prob_w9.json",
    "results/approach_windowed/step2rank_corr_results_rep_velocity_w9.json",
]

tr_rep_paths = [
    "results/approach_windowed/step2threshold_results_rep_note_w9.json",
    "results/approach_windowed/step2threshold_results_rep_finger_nakamura_w9.json",
    "results/approach_windowed/step2threshold_results_rep_finger_w9.json",
    "results/approach_windowed/step2threshold_results_rep_prob_w9.json",
    "results/approach_windowed/step2threshold_results_rep_velocity_w9.json",
]

ws_c_paths = [
    "results/approach_windowed/step2rank_corr_results_rep_velocity_w1.json",
    "results/approach_windowed/step2rank_corr_results_rep_velocity_w3.json",
    "results/approach_windowed/step2rank_corr_results_rep_velocity_w5.json",
    "results/approach_windowed/step2rank_corr_results_rep_velocity_w9.json",
    "results/approach_windowed/step2rank_corr_results_rep_velocity_w13.json",
    "results/approach_windowed/step2rank_corr_results_rep_velocity_w19.json",
]


ws_paths = [
    "results/approach_windowed/results_rep_velocity_w1.json",
    "results/approach_windowed/results_rep_velocity_w3.json",
    "results/approach_windowed/results_rep_velocity_w5.json",
    "results/approach_windowed/results_rep_velocity_w9.json",
    "results/approach_windowed/results_rep_velocity_w13.json",
    "results/approach_windowed/results_rep_velocity_w19.json",
]

tr_ws_paths = [
    "results/approach_windowed/step2threshold_results_rep_velocity_w1.json",
    "results/approach_windowed/step2threshold_results_rep_velocity_w3.json",
    "results/approach_windowed/step2threshold_results_rep_velocity_w5.json",
    "results/approach_windowed/step2threshold_results_rep_velocity_w9.json",
    "results/approach_windowed/step2threshold_results_rep_velocity_w13.json",
    "results/approach_windowed/step2threshold_results_rep_velocity_w19.json",
]

ws_c_paths_prob = [
    "results/approach_windowed/step2rank_corr_results_rep_prob_w1.json",
    "results/approach_windowed/step2rank_corr_results_rep_prob_w3.json",
    "results/approach_windowed/step2rank_corr_results_rep_prob_w5.json",
    "results/approach_windowed/step2rank_corr_results_rep_prob_w9.json",
    "results/approach_windowed/step2rank_corr_results_rep_prob_w13.json",
    "results/approach_windowed/step2rank_corr_results_rep_prob_w19.json",
]


ws_paths_prob = [
    "results/approach_windowed/results_rep_prob_w1.json",
    "results/approach_windowed/results_rep_prob_w3.json",
    "results/approach_windowed/results_rep_prob_w5.json",
    "results/approach_windowed/results_rep_prob_w9.json",
    "results/approach_windowed/results_rep_prob_w13.json",
    "results/approach_windowed/results_rep_prob_w19.json",
]

tr_ws_paths_prob = [
    "results/approach_windowed/step2threshold_results_rep_prob_w1.json",
    "results/approach_windowed/step2threshold_results_rep_prob_w3.json",
    "results/approach_windowed/step2threshold_results_rep_prob_w5.json",
    "results/approach_windowed/step2threshold_results_rep_prob_w9.json",
    "results/approach_windowed/step2threshold_results_rep_prob_w13.json",
    "results/approach_windowed/step2threshold_results_rep_prob_w19.json"
]

gru_paths = [
    "results/approach_attention/results_rep_note.json",
    "results/approach_attention/results_rep_finger.json",
    "results/approach_attention/results_rep_finger_nakamura.json",
    "results/approach_attention/results_rep_prob.json",
    "results/approach_attention/results_rep_velocity.json"
]

gru_rank_paths = [
    "results/approach_attention/rank_results_rep_note.json",
    "results/approach_attention/rank_results_rep_finger.json",
    "results/approach_attention/rank_results_rep_finger_nakamura.json",
    "results/approach_attention/rank_results_rep_prob.json",
    "results/approach_attention/rank_results_rep_velocity.json"
]




def fill_results_p(result_paths):
    results = {}

    for path in result_paths:
        data2 = load_json(path)
        parameters2, pearson_bartok, pearson_henle, split_numbers2 = [], [], [], []
        name = os.path.basename(path)[:-5] 
        for key, d in data2.items():
            split_numbers2.append(int(key))
            pearson_bartok.append(d["pearson_bartok"][0])
            pearson_henle.append(d["pearson_henle"][0])
        results[name] = {
            "pearson_bartok": pearson_bartok,
            "pearson_henle": pearson_henle
        }
    return results

def fill_results(result_paths):
    results = {}

    for path in result_paths:
        data2 = load_json(path)
        parameters2, test_balanced_acc2, train_balanced_acc2, split_numbers2 = [], [], [], []
        name = os.path.basename(path)[:-5] 
        for key, d in data2.items():
            split_numbers2.append(int(key))
            test_balanced_acc2.append(d['test_balanced_acc'])
            train_balanced_acc2.append(d['train_balanced_acc'])
        results[name] = {
            'split_numbers': split_numbers2,
            'train': train_balanced_acc2,
            'test': test_balanced_acc2
        }
    return results

def fill_results_gru(result_paths):
    results = {}

    for path in result_paths:
        data2 = load_json(path)
        parameters2, test_balanced_acc2, train_balanced_acc2, split_numbers2 = [], [], [], []
        name = os.path.basename(path)[:-5] 
        for key, d in data2.items():
            test_balanced_acc2.append(d['test_balanced_acc'])
            train_balanced_acc2.append(d['train_balanced_acc'])
        results[name] = {
            'train': train_balanced_acc2,
            'test': test_balanced_acc2
        }
    return results

def fill_results_c(result_paths):
    results = {}

    for path in result_paths:
        data2 = load_json(path)
        sp_bartok, kb_bartok, kc_bartok, sp_henle, kb_henle, kc_henle = [], [], [], [], [], []
        name = os.path.basename(path)[:-5] 
        for key, d in data2.items():
            sp_bartok.append(d['sp_bartok'][0])
            kb_bartok.append(d['kb_bartok'][0])
            kc_bartok.append(d['kc_bartok'][0])
            sp_henle.append(d['sp_henle'][0])
            kb_henle.append(d['kb_henle'][0])
            kc_henle.append(d['kc_henle'][0])
        results[name] = {
            'sp_bartok': sp_bartok,
            'kb_bartok': kb_bartok,
            'kc_bartok': kc_bartok,
            'sp_henle': sp_henle,
            'kb_henle': kb_henle,
            'kc_henle': kc_henle,
        }
    return results

rep_results = fill_results(rep_paths)
tr_rep_results = fill_results(tr_rep_paths)
ws_results = fill_results(ws_paths)
tr_ws_results = fill_results(tr_ws_paths)

ws_results_prob = fill_results(ws_paths_prob)
tr_ws_results_prob = fill_results(tr_ws_paths_prob)
ws_c_results_prob = fill_results_c(ws_c_paths_prob)

baseline_path = ["results/approach_baseline/results_w9.json"]
baseline_rank_path = ["results/approach_baseline/rank_w9.json"]
baseline_results = fill_results(baseline_path)
baseline_rank_results = fill_results_c(baseline_rank_path)


rep_c_results = fill_results_c(rep_c_paths)
ws_c_results = fill_results_c(ws_c_paths)

rep_gru_results = fill_results_gru(gru_paths)
rep_c_gru_results = fill_results_c(gru_rank_paths)

# baseline approach 
rep_c_results = fill_results_c(rep_c_paths)
ws_c_results = fill_results_c(ws_c_paths)
```

```python

def r(x):
    return np.round(x,2)

def print_results(results):
    for k, v in results.items():
        print(f"REP: {k} | train {r(np.mean(v['train']))}({r(np.std(v['train']))}) | test {r(np.mean(v['test']))}({r(np.std(v['test']))})")
    print()

def print_results_p(results):
    for k, v in results.items():
        print(f"REP: {k} | pearson_bartok {np.mean(v['pearson_bartok'])} | pearson_henle {np.mean(v['pearson_henle'])}")
    print()
    
def print_results_c(results):
    for k, v in results.items():
        for type_corr in ['sp']:
            print(f"REP: {k} | {type_corr}_bartok {r(np.mean(v[f'{type_corr}_bartok']))}({r(np.std(v[f'{type_corr}_bartok']))}) | {type_corr}_henle {r(np.mean(v[f'{type_corr}_henle']))}({r(np.std(v[f'{type_corr}_henle']))})")
    print()

print("the results are presented as mean(stdev), for exaple: 0.79(0.2)")
print("="*40)
print("StepsAlg | The representation comparison results")
print("-"*40)
print("step 1")
print_results(rep_results)
print("step 2 threshold")
print_results(tr_rep_results)
print("step 2 ranking")
print_results_c(rep_c_results)
print("="*40)
print("2StepsAlg | The Window comparison results")
print("-"*40)
print("step 1")
print_results(ws_results)
print("step 2 threshold")
print_results(tr_ws_results)
print("step 2 ranking")
print_results_c(ws_c_results)

print("="*40)
print("2StepsAlg | The Window comparison results NAKAMURA")
print("-"*40)
print("step 1")
print_results(ws_results_prob)
print("step 2 threshold")
print_results(tr_ws_results_prob)
print("step 2 ranking")
print_results_c(ws_c_results_prob)

print("="*40)
print("The accuracy GRU results")
print_results(rep_gru_results)
print("the Ranking GRU")
print_results_c(rep_c_gru_results)

```

```python
def plot_overall(results, title):
    
    layout = go.Layout(
        legend=dict(
            x=0.70,
            y=0.0,
            traceorder='normal',
            font=dict(
                size=12,),
        )
    )
    fig = make_subplots(rows=len(results.keys()), cols=1)

    # Add traces
    for idx, (k, v) in enumerate(results.items()):

        fig.add_trace(go.Scatter(x=v['split_numbers'], 
                                 y=v['train'],
                                 mode='markers',
                                 name=f"{k} train"),
                                 row=idx+1,
                                 col=1
        )
        fig.add_trace(go.Scatter(x=v['split_numbers'], 
                        y=v['test'],
                        mode='markers',
                        name=f"{k} test"),
                        row=idx + 1,
                        col=1
        )
    fig.update_xaxes(tickangle=45)
    fig.update_layout(
        title=title
    )
    fig.show()

plot_overall(rep_results, "step1 representation's comparison")
```

```python
plot_overall(ws_results, "step1 windows' comparison")
```

```python
def plot_box(results, title):
    fig = go.Figure()
    # Add traces
    for idx, (k, v) in enumerate(results.items()):
        fig.add_trace(go.Box(y=v['train'], name=f"{k} (train)", boxpoints="all"))
        fig.add_trace(go.Box(y=v['test'], name=f"{k} (test)", boxpoints="all"))
    
    fig.update_layout(
        title=title
    )
    fig.update_xaxes(tickangle=45)
    fig.show()
    
plot_box(rep_results, "step1 representation's comparison")
plot_box(ws_results, "step1 windows's comparison")
```
